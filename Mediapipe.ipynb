```python
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
import numpy as np
```

```python
model=load_model("maskdetector.model")
```

```python
mpFaceDetection=mp.solutions.face_detection
mpDraw=mp.solutions.drawing_utils
faceDetection=mpFaceDetection.FaceDetection(0.75)
```

```python
capture=cv2.VideoCapture(0)
face=[]
f=[]
font=cv2.FONT_HERSHEY_COMPLEX
while True:
    flag,img=capture.read()
    if flag:
        imgRGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        result=faceDetection.process(imgRGB)
        #print(result)
        
        if result.detections:
            for id,detection in enumerate(result.detections):
                # mpDraw.draw_detection(img,detection)
                bboxC=detection.location_data.relative_bounding_box
                height,width,channel=img.shape
                bbox=int(bboxC.xmin * width),int(bboxC.ymin * height),int(bboxC.width * width),int(bboxC.height * width)
                cv2.rectangle(img,bbox,(255,0,255),2)
                face.append(cv2.resize(imgRGB, (224, 224)))
                face[0] = img_to_array(face[0])
                face[0] = preprocess_input(face[0])
#                 print(face)
                f.append(face)
                face=[]
#                 f = np.array(f, dtype="float32")
#                 f.shape()
                preds = model.predict(f, batch_size=32)
                mask,withoutMask=preds[0][0],preds[0][1]
                #withoutMask = preds[0][1]
                val=detection.score[0]
                if mask > withoutMask:
                    label="Mask"+str(val)
                else:
                    label="No Mask "+str(val)
                #label = "Mask" if mask > withoutMask else "No Mask"
                cv2.putText(img,label,(bbox[0],bbox[0]-20),font,1,(244,250,250),2)
                print(" Prediction is : ",mask," ",withoutMask," ",label)
                cv2.putText(img,"Face Detected ",(20,70),cv2.FONT_HERSHEY_PLAIN,3,(255,0,0),2)
                
        cv2.imshow("Image ",img)
        if cv2.waitKey(2)==27:
            break
capture.release()
cv2.destroyAllWindows()
```
