```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import AveragePooling2D
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np
from imutils import paths
import matplotlib.pyplot as plt
import os
import cv2
```

```python
img=cv2.imread('Dataset/with_mask/1.jpg')
plt.imshow(img)
```

```python
DIRECTORY = "Dataset"
CATEGORIES=["with_mask","without_mask"]
```

```python
data=[]
labels=[]

for c in CATEGORIES:
  count=0
  path=os.path.join(DIRECTORY,c)  
  for img in os.listdir(path):
      img_path=os.path.join(path,img)
      image=load_img(img_path,target_size=(224,224))
      image=img_to_array(image)
      image=preprocess_input(image)
      print(" Image ",count," ",img_path)
      count=count+1
      if count>2000:
            break
      data.append(image)
      labels.append(c)
```

```python
lb=LabelBinarizer()
labels=lb.fit_transform(labels)
labels=to_categorical(labels)
```

```python
data=np.array(data,dtype="float32")
labels=np.array(labels)
```

```python
trainX,testX,trainY,testY=train_test_split(data,labels,test_size=0.20,stratify=labels,random_state=42)
```

```python
INTL_LR=1e-4
EPOCHS=10
BS=32
```

```python
aug= ImageDataGenerator(
    
    rotation_range=20,
    zoom_range=0.15,
    width_shift_range=0.2,
    height_shift_range=0.15,
    shear_range=0.15,
    horizontal_flip=True,
    fill_mode="nearest"
)
```

```python
baseModel=MobileNetV2(weights="imagenet",include_top=False,input_tensor=Input(shape=(224,224,3)))
```

```python
headModel=baseModel.output
headModel=AveragePooling2D(pool_size=(7,7))(headModel)
headModel=Flatten(name="flatten")(headModel)
headModel=Dense(128,activation="relu")(headModel)
headModel=Dropout(0.5)(headModel)
headModel=Dense(2,activation="softmax")(headModel)
```

```python
model=Model(inputs=baseModel.input,outputs=headModel)
```

```python
for layer in baseModel.layers:
    layer.trainable=False
```

```python
print(" Compiling Model ")

opt=Adam(lr=INTL_LR,decay=INTL_LR/EPOCHS)
model.compile(loss="binary_crossentropy",optimizer=opt,metrics=["accuracy"])
```

```python
print(" Training Model ")

H=model.fit(
    aug.flow(trainX,trainY,batch_size=BS),
    steps_per_epoch=len(trainX)//BS,
    validation_data=(testX,testY),
    validation_steps=len(testX)//BS,
    epochs=EPOCHS
)
```

```python
predIdxs = model.predict(testX,batch_size=BS)
predIdxs=np.argmax(predIdxs,axis=1)
```

```python
print(classification_report(testY.argmax(axis=1),predIdxs,target_names=lb.classes_))
```

```python
model.save("maskdetector.model")
```

```python
N=EPOCHS
plt.style.use("ggplot")
plt.figure()
plt.plot(np.arange(0,N),H.history["loss"],Label="train_loss")
plt.plot(np.arange(0,N),H.history["val_loss"],Label="val_loss")
plt.plot(np.arange(0,N),H.history["acc"],Label="train_accuracy")
plt.plot(np.arange(0,N),H.history["val_acc"],Label="val_accuracy")
plt.title("Training Loss and Accuracy ")
plt.xlabel(" Epoch #")
plt.ylabel(" Loss/Accuracy ")
#plt.legend(Loc="lower left")
plt.savefig("plot.png")
```

```python
model.evaluate(testX,testY)
```
