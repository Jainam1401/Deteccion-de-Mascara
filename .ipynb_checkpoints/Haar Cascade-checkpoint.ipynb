```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
```

```python
haar_data=cv2.CascadeClassifier('data.xml')
hd=cv2.CascadeClassifier('haarcascade_eye.xml')
model=load_model("maskdetector.model")
```

```python
capture=cv2.VideoCapture(0)
f=[]
face=[]
count=0
font=cv2.FONT_HERSHEY_COMPLEX
while True:
    flag,img=capture.read()
    if flag:
        faces=haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),4)
            face.append(img[y:y+h,x:x+w,:])
            #plt.imshow(face)
            print("Detecting Face ")
            face[0] = cv2.cvtColor(face[0], cv2.COLOR_BGR2RGB)
            face[0] = cv2.resize(face[0], (224, 224))
            face[0] = img_to_array(face[0])
            face[0] = preprocess_input(face[0])
            f.append(face)
            face=[]
            preds = model.predict(f, batch_size=32)
            mask=preds[0][0]
            withoutMask = preds[0][1]
            label = "Mask" if mask > withoutMask else "No Mask"
            cv2.putText(img,label,(x,y),font,1,(244,250,250),2)
            print(" Prediction is : ",preds," ",label)
        cv2.imshow('Live Image',img)
        if cv2.waitKey(2) == 27:
            break
capture.release()
cv2.destroyAllWindows()
```
